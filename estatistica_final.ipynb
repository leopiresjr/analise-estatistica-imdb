## 2.3 Dicionário das Variáveis

A seguir, descrição das variáveis presentes no conjunto de dados:

- **Poster_Link**: Link da imagem do pôster do filme.

- **Series_Title**: Título do filme.

- **Released_Year**: Ano de lançamento do filme.

- **Certificate**: Classificação indicativa do filme (por exemplo: livre, 12 anos, 16 anos, etc.).

- **Overview**: Breve descrição ou sinopse do filme.

- **Meta_score**: Nota atribuída por críticos especializados (Metacritic).

- **Director**: Nome do diretor do filme.

- **Star1, Star2, Star3, Star4**: Principais atores do elenco.

- **No_of_Votes**: Número total de votos recebidos pelo filme no IMDb.

- **Gross**: Receita bruta do filme.

# 5. Análise Bivariada

## 5.1 Correlação entre variáveis numéricas

A análise de correlação permite avaliar a relação entre variáveis numéricas, indicando o grau de associação entre elas.

### IMDB_Rating × Meta_score

A correlação entre IMDB_Rating e Meta_score é de aproximadamente 0,27, indicando uma relação positiva fraca.

Isso sugere que, embora exista alguma associação entre as avaliações do público e da crítica, elas não estão fortemente alinhadas.

### IMDB_Rating × No_of_Votes

A correlação entre IMDB_Rating e No_of_Votes é de aproximadamente 0,49, indicando uma relação positiva moderada.

Isso sugere que filmes mais bem avaliados tendem a receber um maior número de votos, indicando maior popularidade.

### Meta_score × No_of_Votes

A correlação entre Meta_score e No_of_Votes é praticamente nula (-0,02), indicando ausência de relação significativa entre essas variáveis.

```python
# Gráfico de dispersão entre IMDB_Rating e No_of_Votes
plt.scatter(df['IMDB_Rating'], df['No_of_Votes'])

plt.title('IMDB Rating vs Número de Votos')
plt.xlabel('Nota IMDb')
plt.ylabel('Número de votos')
plt.show()
```

No entanto, essa relação não é perfeitamente linear, sendo possível observar uma dispersão considerável dos pontos.

Além disso, há presença de valores extremos, com alguns filmes apresentando número de votos muito elevado em comparação aos demais.

Essa análise reforça a correlação moderada observada anteriormente entre essas variáveis.

### 6.5 Existe diferença entre a média das avaliações do público (IMDb) e da crítica (Meta_score)?

```python
# Calculando as médias
media_imdb = df['IMDB_Rating'].mean()
media_meta = df['Meta_score'].mean()

media_imdb, media_meta
```

A média das avaliações do público (IMDb) é aproximadamente **7,95**, enquanto a média das avaliações da crítica (Meta_score) é aproximadamente **77,97**.

Apesar das escalas serem diferentes (IMDb varia de 0 a 10, enquanto o Meta_score varia de 0 a 100), é possível observar que ambas indicam avaliações positivas.

No entanto, análises anteriores mostraram que a correlação entre essas variáveis é fraca, indicando que a opinião do público e da crítica nem sempre estão alinhadas.

Isso sugere que filmes bem avaliados pelo público podem não receber a mesma avaliação por parte dos críticos, e vice-versa.

### 6.6 Os filmes mais votados possuem, em média, notas maiores?

```python
# Separando filmes mais votados (acima da mediana)
mediana_votos = df['No_of_Votes'].median()

mais_votados = df[df['No_of_Votes'] > mediana_votos]
menos_votados = df[df['No_of_Votes'] <= mediana_votos]

# Calculando médias
media_mais_votados = mais_votados['IMDB_Rating'].mean()
media_menos_votados = menos_votados['IMDB_Rating'].mean()

media_mais_votados, media_menos_votados
```

Os filmes mais votados apresentam uma média de nota de aproximadamente **8,00**, enquanto os menos votados possuem média de aproximadamente **7,89**.

Essa diferença indica que filmes com maior número de votos tendem a ter avaliações ligeiramente superiores.

Esse resultado reforça a relação positiva moderada observada anteriormente entre número de votos e nota IMDb, sugerindo que popularidade e qualidade podem estar associadas, embora não de forma determinante.
